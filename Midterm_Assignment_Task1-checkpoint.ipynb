**a) Implement the standard matrix multiplication algorithm using Python
 without using any pre-built matrix multiplication functions (like those in
 NumPy).**

```python
def naive_matrix_multiplication(A, B):
    n = len(A)
    #Leere n x n Matrix für das Ergebnis
    C = [[0] * n for _ in range(n)]
    #Produkt berechnen
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C
```

```python
#Beispielmatrizen
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]

#Aufruf der Funktion und Ausgabe des Ergebnisses
result = naive_matrix_multiplication(A, B)
print("Result of naive matrix multiplication:")
for row in result:
    print(row)
```

**b) Analyze and benchmark the runtime of your algorithm on each
 matrix size. Use the Python time module to record the time taken for each
 matrix size.**

```python
import time
import random

#Funktion zur Generierung einer n x n-Matrix mit zufälligen Werten
def generate_random_matrix(n):
    return [[random.randint(1, 10) for _ in range(n)] for _ in range(n)]

sizes = [10, 50, 100, 200]  # Du kannst mehr Größen hinzufügen
naive_times = []

#Benchmark der Laufzeit
for size in sizes:
    A = generate_random_matrix(size)
    B = generate_random_matrix(size)
    
    start_time = time.time()
    naive_matrix_multiplication(A, B)
    end_time = time.time()
    
    time_taken = end_time - start_time
    naive_times.append(time_taken)
    print(f"Size {size} took {time_taken:.4f} seconds")
```

**c) Compare your implementation with the built-in matrix multiplication
 in NumPy. Find any preferred literature to get familiarized with the concept
 of (algorithmic) time complexity. Plot the time complexity for both methods
 as the matrix size increases.**

```python
import numpy as np
import matplotlib.pyplot as plt

numpy_times = []

for size in sizes:
    A = np.random.randint(1, 10, size=(size, size))
    B = np.random.randint(1, 10, size=(size, size))
    
    start_time = time.time()
    np.dot(A, B)
    end_time = time.time()
    
    numpy_times.append(end_time - start_time)
    print(f"NumPy Size {size} took {end_time - start_time:.4f} seconds")

#Plot der Ergebnisse
plt.plot(sizes, naive_times, label="Naive Multiplication")
plt.plot(sizes, numpy_times, label="NumPy Multiplication")
plt.xlabel("Matrix Size")
plt.ylabel("Time (seconds)")
plt.legend()
plt.title("Comparison of Naive vs NumPy Matrix Multiplication")
plt.show()
```

### Zeitkomplexität der Matrizenmultiplikation

**Naive Matrizenmultiplikation:**
- Die naive Implementierung verwendet drei verschachtelte Schleifen (über `i`, `j` und `k`), wodurch jede Kombination von Elementen in der ersten Matrix mit denen der zweiten Matrix multipliziert und addiert wird.
- Da jede der Schleifen von 0 bis n-1 läuft, führt dies führt zu einer Zeitkomplexität von: $$O(n^3)$$

**NumPy Matrizenmultiplikation (`np.dot`):**
- Die `np.dot`-Funktion in NumPy ist hoch optimiert und nutzt effiziente lineare Algebra-Bibliotheken (wie BLAS oder LAPACK). Sie ist auf Geschwindigkeit optimiert und kann zudem Mehrkernprozessoren nutzen.
- Obwohl die theoretische Zeitkomplexität für große Matrizen die gleiche wie bei der naiven Matrizenmultiplikation ist, sind die konstanten Faktoren stark reduziert, was die Berechnung in der Praxis viel schneller macht als bei einer einfachen dreifachen Schleife.

**Vergleich der beiden Methoden:**
- Der Unterschied in der Performance zwischen der naiven Implementierung und der `np.dot`-Funktion zeigt sich insbesondere bei größeren Matrizen. Die Optimierungen in NumPy verkürzen die benötigte Rechenzeit erheblich.


**e) Implement a simple version of Strassen’s algorithm for two-dimensional
 matrices (with and without the recursion) and compare its performance with
 your na¨ ıve approach and your implementation with NumPy. You can use any
 reference of your choice to get familiarized with Strassen’s algorithm.?**

```python
def strassen_matrix_multiplication(A, B):
    #Voraussetzung: Die Matrizen A und B sind quadratisch und die Dimensionen sind Potenzen von 2
    n = len(A)
    if n == 1:
        return [[A[0][0] * B[0][0]]]
    else:
        #Matrix halbieren
        mid = n // 2
        A11 = [[A[i][j] for j in range(mid)] for i in range(mid)]
        A12 = [[A[i][j] for j in range(mid, n)] for i in range(mid)]
        A21 = [[A[i][j] for j in range(mid)] for i in range(mid, n)]
        A22 = [[A[i][j] for j in range(mid, n)] for i in range(mid, n)]
        B11 = [[B[i][j] for j in range(mid)] for i in range(mid)]
        B12 = [[B[i][j] for j in range(mid, n)] for i in range(mid)]
        B21 = [[B[i][j] for j in range(mid)] for i in range(mid, n)]
        B22 = [[B[i][j] for j in range(mid, n)] for i in range(mid, n)]

        #Strassen Matrizen Produkte berechnen
        M1 = strassen_matrix_multiplication(add_matrices(A11, A22), add_matrices(B11, B22))
        M2 = strassen_matrix_multiplication(add_matrices(A21, A22), B11)
        M3 = strassen_matrix_multiplication(A11, subtract_matrices(B12, B22))
        M4 = strassen_matrix_multiplication(A22, subtract_matrices(B21, B11))
        M5 = strassen_matrix_multiplication(add_matrices(A11, A12), B22)
        M6 = strassen_matrix_multiplication(subtract_matrices(A21, A11), add_matrices(B11, B12))
        M7 = strassen_matrix_multiplication(subtract_matrices(A12, A22), add_matrices(B21, B22))

        #C Submatrizen berechnen
        C11 = add_matrices(subtract_matrices(add_matrices(M1, M4), M5), M7)
        C12 = add_matrices(M3, M5)
        C21 = add_matrices(M2, M4)
        C22 = add_matrices(subtract_matrices(add_matrices(M1, M3), M2), M6)

        #Endergebnis zusammenfügen
        new_matrix = [[0] * n for _ in range(n)]
        for i in range(mid):
            for j in range(mid):
                new_matrix[i][j] = C11[i][j]
                new_matrix[i][j + mid] = C12[i][j]
                new_matrix[i + mid][j] = C21[i][j]
                new_matrix[i + mid][j + mid] = C22[i][j]
        return new_matrix

#Hilfsfunktionen für Addition und Subtraktion von Matrizen
def add_matrices(A, B):
    n = len(A)
    return [[A[i][j] + B[i][j] for j in range(n)] for i in range(n)]

def subtract_matrices(A, B):
    n = len(A)
    return [[A[i][j] - B[i][j] for j in range(n)] for i in range(n)]
```

```python
strassen_times = []
sizes = [8, 16, 32, 64] 

for size in sizes:
    A = generate_random_matrix(size)
    B = generate_random_matrix(size)

    start_time = time.time()
    result = strassen_matrix_multiplication(A, B)
    end_time = time.time()

    time_taken = end_time - start_time
    strassen_times.append(time_taken)
    
    print(f"Strassen Size {size} took {time_taken:.4f} seconds")
```

```python
print("Länge von sizes:", len(sizes))
print("Länge von naive_times:", len(naive_times))
print("Länge von numpy_times:", len(numpy_times))
print("Länge von strassen_times:", len(strassen_times))
```
